## Oznake
Lepo je, če na grafu označimo osi.

````python
plt.plot(testi)
plt.plot(pozitivni)
plt.plot(smrti)

plt.xlabel("Datum")
plt.ylabel("Število")

plt.show()
````

Lepo je tudi, če vemo kaj posamezna črta prikazuje.

````python
plt.plot(testi)
plt.plot(pozitivni)
plt.plot(smrti)

plt.xlabel("Datum")
plt.ylabel("Število")

plt.legend(["testi", "pozitivni", "smrti"])

plt.show()
````

Spremenimo lahko način in barvo izrisa, debelino linije...
* "...": tip linije ("-", "--", ".", "o"...)
* linewidth
* color

````python
plt.plot(testi)
plt.plot(pozitivni)
plt.plot(smrti, ".")

plt.xlabel("Datum")
plt.ylabel("Število")

plt.legend(["testi", "pozitivni", "smrti"])

f = plt.gcf() # get current figure
f.set_size_inches(10,5) # nastavi velikost slike (sirina, visina)

plt.show()
````

Še vedno ne vemo, kaj predstavlja npr. 35 na osi x... Uporabimo funkcijo [```xticks```](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.xticks.html)

Najprej pripravimo oznake in njihove lokacije. Vzemimo npr. vsako 30. oznako in ohranimo samo podatek o mesecu.

````python
# lokacije
locs = range(0, len(datumi), 30)
oznake = datumi[::30]

for i in range(len(oznake)):
    oznake[i] = oznake[i].split("-")[1] # pustimo samo mesec
````

Zdaj oznake in lokacije uporabimo pri klicu metode `xticks`.

````python
plt.plot(testi)
plt.plot(pozitivni)
plt.plot(smrti, ".")

plt.xlabel("Mesec")
plt.ylabel("Število")

plt.legend(["testi", "pozitivni", "smrti"])
plt.xticks(locs, oznake)


f = plt.gcf() # get current figure
f.set_size_inches(10,5) # nastavi velikost slike (sirina, visina)


plt.show()
````

# NumPy in CSV
Podatke bi sicer lahko prebrali tudi v ```numpy``` strukturo ```ndarray```. Uporabimo lahko metodo ```np.genfromtxt```.

````python
import numpy as np
````

````python
podatki = np.genfromtxt("stats.csv", delimiter=",", dtype=str)
podatki.shape
glava = podatki[0]
podatki = podatki[1:]
````

Vzamemo ven glavo, `datume` (1)  in podatke: `testi` (4), `pozitivni` (6) in `smrti` (34). Pretvoriti jih moramo v numerično obliko, potem pa jih lahko izrišemo.

````python
podatki = podatki[:,[1, 4, 6, 34]] # vseh vrstic 1., 4., 6., 34. stolpec
datumi = podatki[:, 0] # vseh vrstic nicti stolpec
podatki = podatki[:, 1:] # vseh vrstic stolpci razen nictega
````

````python
podatki[podatki == ""] = np.nan
````

````python
podatki = podatki.astype(float)
````

````python
plt.plot(podatki)
plt.legend(glava[[4,6,34]]) # za indeksiranje uporabimo kar array/list
plt.xticks(range(0, len(datumi), 30), datumi[::30])
plt.show()
````

Problem: Glavo moramo obravnavati posebej, podatke o datumih moramo obravnavati posebej, izvajati moramo pretvorbe v števila, šteti stolpce, ki so za nas zanimivi, itd. Še vedno imamo kar nekaj komplikacij.
