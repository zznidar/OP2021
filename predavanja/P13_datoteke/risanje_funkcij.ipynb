# Risanje funkcij
Narisati želimo potek logaritemskih funkcij $log_{e}(x)$, $log_{10}(x)$ in $log_{2}(x)$ v razponu $x \in [0,5]$

Najprej moramo uvoziti potrebne module:

````python
import matplotlib.pyplot as plt
import numpy as np
````

Potem napišimo funkcijo, ki bo sprejela seznam vrednosti na osi x in 3 sezname vrednosti na osi y, vse skupaj izrisala in opremila z ustreznimi oznakami

````python
def risi(X, Y1, Y2, Y3, oznake): # dejmo probat (X, *rest)
    plt.plot(X, Y1)
    plt.plot(X, Y2)
    plt.plot(X, Y3)
    
    plt.legend(oznake)
    plt.show()
````

````python
def risi2(X, *Y, oznake):
    for y in Y:
        plt.plot(X, y)
    plt.legend(oznake)
    plt.show()
    # tu so oznake keyword-only argument
````

Za tabeliranje vrednosti bomo uporabili modul ```numpy``` in njegove funkcije za izračun logaritmov.

````python
X = np.arange(0.01, 5.01, 0.01)
Y_e = np.log(X)
Y_10 = np.log10(X)
Y_2 = np.log2(X)

risi(X, Y_e, Y_10, Y_2, ["$log_{e}(x)$", "$log_{10}(x)$", "$log_{2}(x)$"])

risi(X, Y_e, Y_10, Y_2, oznake=["$log_{e}(x)$", "$log_{10}(x)$", "$log_{2}(x)$", "test"])
````
