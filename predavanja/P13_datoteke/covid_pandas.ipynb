### Preimenovanje stolpcev
Za začetek bi bilo lažje, če bi imeli stolpci krajša imena. Stolpce lahko preimenujemo z nastavitvijo lastnosti ```columns```

````python
df2.columns
````

````python
df2.columns = ["datumi", "datumi", "kek", 3]
````

````python
df2["datumi"]
````

## Spreminjanje vrednosti `nan`
Spremenimo vse vrednosti `nan` v 0 (glede na podatke, s katerimi delamo, je to v tem primeru pravilno). Uporabimo lahko metodo `fillna`. Vstavljanje se vrši na kopiji tabele.

````python
df2 = df2.fillna(0)
````

### Indeksiranje 
Do posameznega stolpca lahko dostopamo kar preko njegovega imena. 

````python
df2[["kek", 3]]
````

Lahko podamo več stolpcev naenkrat, katerih imena zapakiramo v seznam

Če kakšnega stolpca ne potrebujemo, ga lahko tudi izbrišemo. Uporabimo metodo ```drop```, ki ji podamo argument ```columns```. Brisanje se vrši na kopiji tabele!

````python
df2.drop(columns=["datumi"])
````

### Indeksiranje z uporabo metode ```loc```
Če želimo izvesti indeksiranje po vrsticah, uporabimo metodo ```loc```. Metodi ```loc```podamo ```index``` vrstice, ki je v našem primeru kar njena zaporedna številka (lahko pa bi bila tudi kaj drugega)

````python
df.loc[0:10, "date"]
````

Metodi ```loc``` lahko podamo tudi stolpec (podobno kot smo indeksirali matriko pri strukturi tipa ```ndarray```)

### Spreminjanje indexa vrstic
Če bi npr. želeli, da ```index```vrstice določa mesec, bi lahko uporabili metodo ```set_index```

````python
df3 = df2.set_index("datumi")
````

Zdaj metodi ```loc``` podamo kar datum, ki nas zanima (zaporedne številke vrstice ne sprejme več)

````python
df3.loc("2022-01-07") 
# to bi delalo, ce ne bi imeli 2 stolpcov "datumi", ker zdej so indeksi terke od dveh stolpcev :sweat_smile
````

Spet ji lahko poleg vrstice podamo tudi stolpec oziroma stolpce, ki jih želimo pridobiti.

### Indeksiranje z uporabo metode ```iloc```
Če želimo izvesti indeksiranje po številkah vrstic (in stolpcev), uporabimo metodo ```iloc```. Metodi ```iloc```podamo zaporedno številko vrstice, lahko pa tudi stolpca (štetje se začne z 0)

````python
df.iloc[10, 2]
````

Poskusimo še s številko stolpca.

Kot pri indeksiranju seznamov lahko tudi tu delamo rezine

````python
df.iloc[:100, 2]
````

Če bi želeli izpisati vse vrstice v določenem stolpcu (po njegovi zaporedni številki), lahko za indeks vrstice uporabimo znak ```:```

### Filtriranje podatkov
Kaj se zgodi, če pri indeksiranju uporabimo primerjalni operator, kot sta npr. ```==``` ali ```>```

````python
df2["kek"] > 2000
````

Izpišimo samo tiste vrstice, ko je bilo več kot 2400 pozitivnih primerov

````python
df2[df2["kek"] > 2000]
````

Kaj pa če bi želeli iz teh vrstic izpisati samo število testov in število pozitivnih?

Primerjalne operatorje lahko tudi združujemo z uporabo operatorja ```&``` (*bitwise and*) in operatorja ```|``` (*bitwise or*), rezultate pa negiramo z uporabo operatorja ```~```(*bitwise not*). 

Izpišimo vrstice, ko je bilo število pozitivnih med 1500 in 2000.

````python
(df2["kek"] > 1500) & (df2["kek"] < 2000)
````

````python
df2.loc[(df2["kek"] > 1500) & (df2["kek"] < 2000), ["kek", 3]]
````
